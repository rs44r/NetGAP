```python
import json
import os

def load_from_file(directory,filename):
    filename = f"{directory}\{filename}.json"
    with open(filename, 'r', encoding='utf-8') as fo:
        serialized_data = json.load(fo)
            
    return  serialized_data

directory="C:/Users\smora\Documents\LiU\Workspace\\NetGAP\data"
filename="saab_usecase_1"
extended_uc=load_from_file(directory,filename)
```

```python
process_list=[p["name"] for p in extended_uc["processes"]]
print(process_list)

message_list=[ (m["from_process"], m["to_process"], m["size"]) for m in extended_uc["communications"]]
print(message_list[0:10])
```

```python
import networkx as nx
G=nx.DiGraph()
G.add_nodes_from(process_list)

for m in message_list:
    G.add_edge(m[0],m[1],weigth=m[2])
```

```python
pos=nx.spring_layout(G)
edges = nx.draw_networkx_edges(G, pos=pos,arrowsize=10)
nodes = nx.draw_networkx_nodes(G, pos=pos,node_size=100)
gexf_name=filename+".gexf"
nx.write_gexf(G, gexf_name)
```
