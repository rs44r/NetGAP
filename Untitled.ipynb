```python
import time
class CustomRng:
    def __init__(self, seed=None):
        self.multiplier = 1103515245
        self.increment = 12345
        self.modulus = 2**32  
        self.calls = 0 
        
        if seed!=None:
            self.seed=seed
        else:
            self.seed=int(time.time_ns())

    def _random_LCG(self):
        self.seed = (self.multiplier * self.seed + self.increment) % self.modulus
        self.calls+=1
        return self.seed
    
    def random(self):
        return self._random_LCG()/self.modulus
    
    def randint(self, a, b):
            if a>=b:
                raise Exception("a>=b")
            return a + self._random_LCG() % (b - a + 1)
```

```python
class Message:
    def __init__(self,mid,src,dst,freq,size):
        self.mid=mid
        self.src=src
        self.dst=dst
        self.freq=freq
        self.size=size
```

```python
import json 
def load_from_file(directory,filename):
    filename = f"{directory}\{filename}.json"
    with open(filename, 'r', encoding='utf-8') as fo:
        serialized_data = json.load(fo)
        message_list_from_file=[x for x in serialized_data["message_list"].values()]
        message_list=[Message(m["mID "],m["source"],m["dest"],m["freq"],m["size"]) for m in message_list_from_file]  
    return message_list  

```

```python
class Individual:
    def __init__(self):
        self.genome=[0 for _ in range(processNumber)]
        self.fitness=0
    def init(self):
            self.genome=[random.randint(0,moduleNumber-1) for _ in range(processNumber)]
    

def cross(p1,p2):
    index = random.randint(0,processNumber-1)
    offspring1=Individual()
    offspring2=Individual()
    
    if p1==p2:
        offspring1.genome[:]=p1.genome[:]
        offspring2.genome[:]=p1.genome[:]
        
    else:
        offspring1.genome[0:index]=p1.genome[0:index]
        offspring1.genome[index:]=p2.genome[index:]

        offspring2.genome[0:index]=p2.genome[0:index]
        offspring2.genome[index:]=p1.genome[index:]
        
    return offspring1, offspring2

def swap(individual,rate):
    
    if random.random()<rate:
        index1=random.randint(0,len(individual.genome)-1)
        index2=random.randint(0,len(individual.genome)-1)

        while index1 == index2 :
            index2=random.randint(0,len(individual.genome)-1)

        individual.genome[index1], individual.genome[index2] = individual.genome[index2], individual.genome[index1]
                                           
def eval(individual):
    genome=individual.genome
    IN=0
    OUT=1
    module_usage=[0 for x in range(moduleNumber)]
    bandwidth_usage=[[0,0] for x in range(moduleNumber)]
    
    for i,v in enumerate(genome):
        module_usage[v]+=process_loads[i]
        
    for i,m in enumerate(message_list):
        #print(m.__dict__)
        src_mod=genome[m.src]
        dst_mod=genome[m.dst]
        
        if src_mod!=dst_mod:
            bandwidth_usage[src_mod][OUT]+=m.freq*m.size
            bandwidth_usage[dst_mod][IN]+=m.freq*m.size
        
    max_computation_load=0
    max_bandwidth_load=0
    for i in range(moduleNumber):
        if module_usage[i]>max_computation_load:
            max_computation_load=module_usage[i]
        if  bandwidth_usage[i][OUT]> max_bandwidth_load:
            max_bandwidth_load=bandwidth_usage[i][OUT]
        if  bandwidth_usage[i][IN]> max_bandwidth_load:
            max_bandwidth_load=bandwidth_usage[i][IN]
    
    #print(max_computation_load)
    #print(max_bandwidth_load)
    
    
    loads={}
    loads.update({"max_computation":max_computation_load})
    loads.update({"max_bandwidth":max_bandwidth_load})
    loads.update({"module_usage": module_usage})
    loads.update({"bandwidth_usage":bandwidth_usage})
    
   
    score=1/((max_computation_load/(moduleCap*0.8))**8+(max_bandwidth_load/(linkCap*0.8))**8)
    individual.fitness=score
    return score,loads
    
def selection(population):
    total_fitness=0
    prob_array=[]
    
    for individual in population:
        total_fitness+=individual.fitness
        
    #print(total_fitness)
    
        
    for _,individual in enumerate(population):
        prob_array.append(individual.fitness/total_fitness)

    def pick():
        r = random.random()
       
        s=0
        for i,p in enumerate(prob_array):
            s+=p
            if s > r:
                #print(s,r,i)
                return i
        return len(population)-1
    
    index1=pick()
    index2=pick()
    return index1,index2
```

```python
import time
def run():
    start=time.time()
    random.seed=12345
    random.calls=0

    generations =300
    pop_size = 300

    population=[Individual() for x in range(pop_size)]
    for ind in population:
        ind.init()
    best_fitness=0

    for i in range(generations):
        for individual in population:
            eval(individual)

        population=sorted(population,key = lambda individual: individual.fitness,reverse=True)
        best_fitness=population[0].fitness

        if best_fitness>0.6:
            print(f"found in {i} f:{best_fitness:.3f}")
            break
            #pass
        else:
            print(f"{i}:{best_fitness:.3f}||",end="")

            new_population=[Individual() for x in range(len(population))]
            for i in range(0):
                new_population[i]=population[i]

            for i in range (0,pop_size,2):
                index1,index2=selection(population[0:int(len(population)/2)])
                #print(index1,index2)
                offspring1,offspring2=cross(population[index1],population[index2])
                new_population[i]=offspring1
                new_population[i+1]=offspring2

            for individual in new_population:
                swap(individual,0.15)

            population=new_population

    duration = time.time()-start
    print(f"terminated: {best_fitness:.3f} -> {duration:.2f}s ")

  
```

```python
import time
import cProfile
import pstats
 
# Run the function with cProfile and print a flame graph
profiler = cProfile.Profile()
profiler.enable()
 
run()
 
profiler.disable()
profiler.dump_stats('test.prof')
!python flameprof.py test.prof > requests.svg
from IPython.display import SVG, display
display(SVG(filename="requests.svg"))
```
